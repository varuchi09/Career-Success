# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier, XGBRegressor
from lightgbm import LGBMClassifier, LGBMRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import train_test_split
from sklearn.base import clone
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, f1_score, mean_absolute_error
import joblib
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

# Feature Engineering

```python
df = df.drop(columns=['student_id'], errors='ignore')

# 1. Success Score: Combining academic and communication excellence
df['Success_Score'] = (df['cgpa'] * df['communication_skills']) / (df['academic_backlogs'] + 1)

# 2. Portfolio Strength: Weighting projects and certifications heavily
df['Portfolio_Strength'] = (df['projects'] * 1.5) + (df['certifications'] * 1.2) + (df['internships'] * 2.0)

# 3. skill_depth: Combined skills total
df['skill_depth'] = df['communication_skills'] + df['certifications'] + df['projects']

print('M+ Sanitized Features engineered successfully.')
df.head()
```

```python
df['skills'] = df['communication_skills'] + df['certifications'] + df['projects']
```

```python
df = df[df['salary_lpa'] < df['salary_lpa'].quantile(0.95)]
```

```python
df.info()
```

# Feature transformation

```python
df['placement_rate'].unique()
```

```python
exclude = ['placed_3m', 'placed_6m', 'placed_12m', 'salary_lpa', 'placement_rate']
num_cols = [c for c in df.drop(columns=exclude).columns if c not in ['branch', 'region', 'internship_quality', 'market_demand']]
print(f'Final Training Features Count: {len(num_cols) + 4}')
```

```python
ct = ColumnTransformer(transformers = [('num', StandardScaler(), num_cols),
                                       ('ohe', OneHotEncoder(drop = 'first', sparse_output = False, handle_unknown = 'ignore'), ['branch', 'region']),
                                       ('oe', OrdinalEncoder(categories = [['none', 'low', 'medium', 'high'],['low','medium','high']]), ['internship_quality', 'market_demand'])
    
])
```

# Pipelines

## i. Classification pipeline

```python
clf_pipe = Pipeline([('ct', ct),
                     ('model', XGBClassifier(n_estimators = 100))])
```

## ii. Regression pipeline

```python
xgb_reg = XGBRegressor(n_estimators=1000, learning_rate=0.01, max_depth=6, subsample=0.8, colsample_bytree=0.8, random_state=42)
lgbm_reg = LGBMRegressor(n_estimators=1000, learning_rate=0.01, max_depth=6, subsample=0.8, importance_type='gain', random_state=42, verbose=-1)

reg_ensemble = VotingRegressor([('xgb', xgb_reg), ('lgbm', lgbm_reg)])
reg_pipe = Pipeline([('ct', ct), ('model', reg_ensemble)])
print('M+ Voting Ensemble Ready.')
```

# Splitting dataset

```python
x = df.drop(columns=['placed_3m', 'placed_6m', 'placed_12m', 'salary_lpa', 'placement_rate'])
y = df[['placed_3m', 'placed_6m', 'placed_12m', 'salary_lpa', 'placement_rate']]
print('X Columns:', x.columns.tolist())
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_3m_train = y_train['placed_3m']
y_6m_train = y_train['placed_6m']
y_12m_train = y_train['placed_12m']
y_sal_train = y_train['salary_lpa']
y_placement_rate_train = y_train['placement_rate']

y_3m_test = y_test['placed_3m']
y_6m_test = y_test['placed_6m']
y_12m_test = y_test['placed_12m']
y_sal_test = y_test['salary_lpa']
y_placement_rate_test = y_test['placement_rate']
```

# Model training

```python
model_3m = clone(clf_pipe)
model_6m = clone(clf_pipe)
model_12m = clone(clf_pipe)
model_sal = clone(reg_pipe)
model_p_rate = clone(reg_pipe)
```

```python
model_3m.fit(x_train, y_3m_train)
model_6m.fit(x_train, y_6m_train)
model_12m.fit(x_train, y_12m_train)
model_sal.fit(x_train, y_sal_train)
model_p_rate.fit(x_train, y_placement_rate_train)
```

```python
y3m_pred = model_3m.predict(x_test)
y6m_pred = model_6m.predict(x_test)
y12m_pred = model_12m.predict(x_test)
y_sal_pred = model_sal.predict(x_test)
y_p_rate_pred = model_p_rate.predict(x_test)
```

# Evaluating models

## classification models

```python
f1_score_y3 = f1_score(y3m_pred, y_3m_test)
f1_score_y6 = f1_score(y6m_pred, y_6m_test)
f1_score_y12 = f1_score(y12m_pred, y_12m_test)

acc_score_y3 = accuracy_score(y3m_pred, y_3m_test)
acc_score_y6 = accuracy_score(y6m_pred, y_6m_test)
acc_score_y12 = accuracy_score(y12m_pred, y_12m_test)
```

## regression models

```python
f1_score_y3 = f1_score(y_3m_test, y3m_pred)
f1_score_y6 = f1_score(y_6m_test, y6m_pred)
f1_score_y12 = f1_score(y_12m_test, y12m_pred)

acc_score_y3 = accuracy_score(y_3m_test, y3m_pred)
acc_score_y6 = accuracy_score(y_6m_test, y6m_pred)
acc_score_y12 = accuracy_score(y_12m_test, y12m_pred)

r2_sal = r2_score(y_sal_test, y_sal_pred)
r2_rate = r2_score(y_placement_rate_test, y_p_rate_pred)

mae_sal = mean_absolute_error(y_sal_test, y_sal_pred)
rmse_sal = np.sqrt(mean_squared_error(y_sal_test, y_sal_pred))
rmse_rate = np.sqrt(mean_squared_error(y_placement_rate_test, y_p_rate_pred))

print(f'Salary R2: {r2_sal:.4f} | R2 Rate: {r2_rate:.4f}')
```

```python
performance_c = pd.DataFrame({'F1 Score' : [f1_score_y3, f1_score_y6, f1_score_y12], 'Accuracy Score' : [acc_score_y3, acc_score_y6, acc_score_y12]}, 
                            index = ['y_3m', 'y_6m', 'y_12m'])
performance_c
```

```python
performance_r = pd.DataFrame({'R2 Score' : [r2_sal, r2_rate], 'RMSE Score' : [rmse_sal, rmse_rate]},
                            index = ['y_salary', 'y_placement_rate'])
performance_r
```
